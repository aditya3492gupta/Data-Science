```python
min_ducks=runs[['Player','0']].sort_values(by='0')
plt.figure(figsize=(10,10))
graph=sn.barplot(y='Player',palette='tab20',x='0',data=min_ducks.head(20))
graph.set_title('Least Ducks By a Batsman')
plt.show()
```

```python
max_ducks=runs[['Player','0']].sort_values(by='0',ascending=False)
plt.figure(figsize=(10,10))
graph=sn.barplot(y='Player',palette='tab10',x='0',data=max_ducks.head(20))
graph.set_title('Most Ducks By a Batsman')
plt.show()
```

```python
max_6=runs[['Player','6s']].sort_values(by='6s',ascending=False)
plt.figure(figsize=(10,10))
graph=sn.barplot(y='Player',palette='tab10',x='6s',data=max_6.head(20))
graph.set_title('Most 6s By a Batsman')
plt.show()
```

```python
max_RUN=runs[['Player','Runs']].sort_values(by='Runs',ascending=False)
plt.figure(figsize=(10,10))
graph=sn.barplot(y='Player',palette='tab20',x='Runs',data=max_RUN.head(20))
graph.set_title('Most Runs By a Batsman')
plt.show()
```
